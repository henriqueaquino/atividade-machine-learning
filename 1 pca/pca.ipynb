```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data = np.random.randn(100, 3)
```

```python
data_mean = np.mean(data, axis=0)
data_centered = data - data_mean

cov_matrix = np.cov(data_centered.T)

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

eigenvectors_subset = eigenvectors[:, np.argsort(eigenvalues)[-2:]]

data_reduced = np.dot(data_centered, eigenvectors_subset)
```

```python
plt.figure(figsize=(12, 5))
```

```python
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], c='blue', edgecolor='k', label="Original Data")
plt.title('Dados Originais (primeiras 2 dimensões)')
plt.xlabel('Dimensão 1')
plt.ylabel('Dimensão 2')
plt.legend()
```

```python
plt.subplot(1, 2, 2)
plt.scatter(data_reduced[:, 0], data_reduced[:, 1], c='red', edgecolor='k', label="Dados Reduzidos (2D)")
plt.title('Dados Após PCA (2 dimensões)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()
```
