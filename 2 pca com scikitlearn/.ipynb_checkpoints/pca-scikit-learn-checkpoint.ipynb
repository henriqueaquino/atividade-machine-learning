```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
X = iris.data 
y = iris.target  
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
print(f"Variância explicada pelo 1º componente: {explained_variance[0]:.2f}")
print(f"Variância explicada pelo 2º componente: {explained_variance[1]:.2f}")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=100)
plt.title('Iris Dataset Reduzido para 2 Componentes Principais')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.colorbar(label='Classe')
plt.grid(True)
plt.show()
```
